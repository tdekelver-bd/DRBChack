restrict tags to the ones that appear in at least 30 docs

```
tagsdf = pd.read_csv('./data/Common_tags_new.csv')
# keep only tags with at least 30 docs
restricted_tags = tagsdf[tagsdf['ntext'] >= 30]
restricted_tags.shape
```

```
tagsdf[tagsdf.ntext == tagsdf.ntext.max()]
```

```
tags = set(restricted_tags.new_tags) - set(['beschikkend gedeeltenieuwe rechtstechnieken'])
# make a set out of it, but remove beschikkend gedeeltenieuwe rechtstechnieken as it appears everywhere
```

```
def tag_in_set(liste):
  out= []
  for el in liste:
    if el in tags:
      out.append(el)
  return out
```

```
train['cleaned_tags'] = train['tags'].apply(tag_in_set)
train['n_tags_cleaned'] = train.cleaned_tags.str.len()
```

```
test = test.append(no_cleaned_tags, ignore_index=True)
```

transform dataframe to be right format of input for model: Simple Transformers requires a column `labels` which contains multi-hot encoded lists of labels as well as a column `text` which contains all the text

```
labels = list(tags)
labels.sort()
```

```
def multi_hot_encode(liste):
  out = [0,] * len(labels) # create empty labels
  for lab in liste:
    idx = labels.index(lab)
    out[idx] = 1
  return out
```

```
train['labels'] = train['cleaned_tags'].apply(multi_hot_encode)
```

# Training
based on [this](https://towardsdatascience.com/multi-label-classification-using-bert-roberta-xlnet-xlm-and-distilbert-with-simple-transformers-b3e0cda12ce5)

```
from simpletransformers.classification import MultiLabelClassificationModel
from sklearn.model_selection import train_test_split
```

```
train_df, eval_df = train_test_split(train, test_size=0.2)
```

```
model = MultiLabelClassificationModel('bert', 'bert-base-dutch-cased', num_labels=len(labels), 
                                      args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 
                                            'num_train_epochs': 10, 'max_seq_length': 512, 'overwrite_output_dir': True})
```

```
model.train_model(train_df)
```

# evaluation

```
result, model_outputs, wrong_predictions = model.eval_model(eval_df)
```
